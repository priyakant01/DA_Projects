```python
import pandas as pd
df = pd.read_csv("dairy_dataset - Copy.csv")
df
```

```python
# infection of Data 
df.info()
```

```python
df["Production Date"] = pd.to_datetime(df["Production Date"])
```

```python
df.head()
```

## Que_1. does number of cows in a farm has any dependency on area aquired. 

```python

import numpy as np
area = np.array(df["Total Land Area (acres)"])
cows = np.array(df["Number of Cows"])
print(np.corrcoef([area,cows]))
# there is no dependancy between  cows and total land area.
```

## 2. what is the average expiry days of each product. 

```python
gb = df.groupby("Product Name").agg({"Shelf Life (days)":"mean"})
gb
```

## Que3. mention the name of the products with max and min shelf life

```python
max_shelf_life = df["Shelf Life (days)"].idxmax()
max_shelf_life
```

```python
df.loc[max_shelf_life]
```

```python
min_shelf_life = df["Shelf Life (days)"].idxmin()
min_shelf_life
```

```python
df.loc[min_shelf_life]
```

## Que4. summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them. 

```python
gb = df.groupby("Brand").agg({"Total Land Area (acres)":"sum", "Number of Cows":"sum","Approx. Total Revenue(INR)":"sum"})
gb
```

## Que5. What is the average shelf life for different storage conditions.

```python
gb = df.groupby(["Brand","Product Name","Storage Condition"]).agg({"Shelf Life (days)":"mean"})
gb
```
